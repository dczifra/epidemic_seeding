```python
gen = SimpleHierarchicalGraph()

graph, per = gen.get(depth = 7, k=3)
print(len(graph))
```

```python
for k in range(3,6):
    graph, per = gen.get(depth = 5, k=k)
    fig,axs = plt.subplots(1,2, figsize=(16,4))
    nx.draw(graph, node_size=30, ax=axs[0])
    
    axs[1].hist(dict(nx.degree(graph)).values(), bins=30)
    axs[1].set_yscale('log')
    plt.show()
```

```python
for k in range(3,6):
    graph, per = gen.get(depth = 3, k=k, mode="cycle")
    fig,axs = plt.subplots(1,2, figsize=(16,4))
    nx.draw(graph, node_size=30, ax=axs[0])
    
    axs[1].hist(dict(nx.degree(graph)).values(), bins=30)
    axs[1].set_yscale('log')
    plt.show()
```

```python
for k in range(3,6):
    graph, per = gen.get(depth = 2, k=k, mode="dense")
    fig,axs = plt.subplots(1,2, figsize=(16,4))
    nx.draw(graph, node_size=30, ax=axs[0])
    
    axs[1].hist(dict(nx.degree(graph)).values(), bins=30)
    axs[1].set_yscale('log')
    plt.show()
```
